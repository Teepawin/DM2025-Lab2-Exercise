# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
import pandas as pd
import json

# Load JSON file
with open("final_posts.json", "r",encoding="utf-8") as f:
    data = json.load(f)

# Flatten the nested JSON structure
df = pd.json_normalize(data, record_path=None)

df
```

```python
# Load data_identification
df_id = pd.read_csv("data_identification.csv")
df_id
```

```python
df_em = pd.read_csv("emotion.csv")
df_em
```

```python
# Merge all columns into one datasets and rename some columns's name

df_posts = df.rename(columns={
    "root._source.post.post_id": "id",
    "root._source.post.text": "text",
    "root._source.post.hashtags": "hashtags"
})[["id", "text", "hashtags"]]

df_merged = df_posts.merge(df_id, on="id", how="left")

df_merged = df_merged.merge(df_em, on="id", how="left")

df_merged
```

```python
# Cleaning data by lowercase, remove URLs, remove mentions, remove extra spaces
import re

def clean_text(t):
    t = t.lower()
    t = re.sub(r"http\S+|www\S+", "", t)  # remove URLs
    t = re.sub(r"@\w+", "", t)           # remove @mentions
    t = re.sub(r"\s+", " ", t).strip()
    return t

df_merged["clean_text"] = df_merged["text"].apply(clean_text)

df_merged
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section

# Combine text + hashtags into text_plus_hashtag

# Make sure all hashtags are lists
df_merged["hashtags_processed"] = df_merged["hashtags"].apply(lambda x: x if isinstance(x, list) else [])

# Convert list of hashtags to a string, empty list becomes empty string
df_merged["hashtags_text"] = df_merged["hashtags"].apply(lambda x: " ".join(x) if x else "")

# Combine clean_text and hashtags_text into one column
df_merged["text_plus_hashtag"] = df_merged["clean_text"] + " " + df_merged["hashtags_text"]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_merged["emotion_id"] = le.fit_transform(df_merged["emotion"])
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label mapping:", label_mapping)
```

```python
# Split the merged dataset into training and test sets
# The "split" column comes from data_identification.csv and tells us
# whether each post belongs to the training set or the test set (no labels).
train_df = df_merged[df_merged["split"] == "train"]
train_df = train_df.drop(columns=["text","split","hashtags","hashtags_processed","clean_text","hashtags_text","emotion"])
# Test set contains posts without emotion labels that we must later predict
test_df  = df_merged[df_merged["split"] == "test"]
test_df = test_df.drop(columns=["text","split","emotion", "hashtags","hashtags_processed", "clean_text","hashtags_text","emotion","emotion_id"])

# Print to verify shapes or preview the data
print(train_df.head())
print(test_df.head())
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    train_df, 
    test_size=0.1, 
    stratify=train_df["emotion_id"], 
    random_state=42
)


import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

tfidf = TfidfVectorizer(max_features=100000, tokenizer=nltk.word_tokenize)

# Fit TF-IDF on training text
X_train = tfidf.fit_transform(train_df["text_plus_hashtag"])
X_val   = tfidf.transform(val_df["text_plus_hashtag"])
```

## 3. Model Implementation Steps

```python
# Add the code related to the model implementation steps in cells inside this section
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Prepare target labels
y_train = train_df["emotion_id"]
y_val   = val_df["emotion_id"]

# max_iter=3000        → allow more optimization steps so model actually converges
# class_weight='balanced' → handle class imbalance by giving rare classes higher weight
# solver='lbfgs'       → supports multinomial logistic regression (softmax)
clf = LogisticRegression(max_iter=10000,class_weight='balanced',solver='lbfgs')
clf.fit(X_train, y_train)

# Evaluate the model
y_pred = clf.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python


# -----------------------------------------
# 2. Transform text using trained TF-IDF
# -----------------------------------------

X_test = tfidf.transform(test_df["text_plus_hashtag"])

# -----------------------------------------
# 3. Predict emotion ID
# -----------------------------------------

test_pred_ids = clf.predict(X_test)

id_to_emotion = {
 0: 'anger',
 1: 'disgust',
 2: 'fear',
 3: 'joy',
 4: 'sadness',
 5: 'surprise'
}

# Convert emotion ID → emotion label
test_pred_labels = [id_to_emotion[i] for i in test_pred_ids]

# Attach to dataframe
test_df["emotion"] = test_pred_labels

# -----------------------------------------
# 4. Save CSV in required output format
# -----------------------------------------

output_df = test_df[["id", "emotion"]]
output_df.to_csv("test_predictions.csv", index=False)

print("Saved as test_predictions.csv")
output_df.head()
```
